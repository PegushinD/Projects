### Проверка корректности данных эффективности обогащения

```python
#Проверим, что эффективность обогащения рассчитана правильно.
#Вычислим её на обучающей выборке для признака rougher.output.recovery
df_train['recovery_count'] = (
    (df_train['rougher.output.concentrate_au']*(df_train['rougher.input.feed_au'] - df_train['rougher.output.tail_au']))/
    (df_train['rougher.input.feed_au']*(df_train['rougher.output.concentrate_au'] - df_train['rougher.output.tail_au']))*100
)
```

```python
recovery = df_train['rougher.output.recovery']
recovery_count = df_train['recovery_count']
```

```python
print('MAE между расчетными и исходными значениями признака:', round(mean_absolute_error(recovery_count,recovery),5))
```

*Расчитали по формуле эффективность обогщения и сравнили показатели с данными которые были в предоставленных датасетах, вычислили среднюю абсолютную ошибку, значение которой очень близко к нулю, а значит коэффициенты эффективномти в исходных данных расчитаны верно*

### Подготовка признаков и целевого показателя для обучения модели

```python
# Так как некоторые параметры недоступны, потому что замеряются и/или рассчитываются значительно позже 
#они в  тестовой выборке отсутствуют,удалим эти признаки из обучающей выборки, кроме целевых. 
target_columns = set(['rougher.output.recovery','final.output.recovery'])
columns_for_del = difference_columns - target_columns
df_train = df_train.drop(columns=columns_for_del, axis=1)
df_train.shape
```

```python
#добавим в тестовую выборку целевые показатели используя дату как ключ
df_test = df_test.merge(df_full[['final.output.recovery', 'rougher.output.recovery', 'date']], on='date')
```

```python
#Установим стлбец 'data'  в качестве индекса для всех выборок
df_train = df_train.set_index('date')
df_test = df_test.set_index('date')
df_full = df_full.set_index('date')
```

```python
# подготовим выборки для обучения моделей
features_train = df_train.drop(['rougher.output.recovery','final.output.recovery'], axis=1)
target_train_rouger =  df_train['rougher.output.recovery']
target_train_final = df_train['final.output.recovery']
```

```python
#напишем функцию для вычисления SMAPE
def get_smape(target, prediction):
    smape = (1/len(target))*np.sum((np.abs(target - prediction))/( (np.abs(target)+np.abs(prediction))/2))*100
    return smape
```

```python
#напишем функцию для вычисления итоговой SMAPE
def get_final_smape(smape_rouger, smape_final):
    final_smape = 0.25 * smape_rouger + 0.75 * smape_final
    return final_smape
```

*На данном этапе мы провели подготовительные действия с выборками для обучения различных моделей. Добавили в тестовую выборку целевые признаки, удадлили из обучающей выборки признаки отсутсвующие в тестовой,разделили обучающую выборку на признаки и целевые показатели, написали функции для расчета целевой метрики SMAPE*

### Обучение модели линейной регрессии

```python
#обучим модель линейной регрессии для предсказания эффективности на стадии флотации
model_rouger_lr = LinearRegression()
smape_scorer = make_scorer(get_smape, greater_is_better=False)
scores_rouger_lr = cross_val_score(model_rouger_lr, features_train, target_train_rouger, scoring = smape_scorer, cv = 5)
smape_rougher = np.mean(scores_rouger_lr)
print('Средняя оценка качества sMape на этапе флотации', round(smape_rougher, 2)*-1)
```

```python
#обучим модель линейной регрессии для предсказания эффективности на финальной стадии
model_final_lr = LinearRegression()
smape_scorer = make_scorer(get_smape, greater_is_better=False)
scores_final_lr = cross_val_score(model_final_lr, features_train, target_train_final, scoring = smape_scorer, cv = 5)
smape_final = np.mean(scores_final_lr)
print('Средняя оценка качества sMape на финальном этапе', round(smape_final, 2)*-1)
```

```python
# получим значения итогового SMAPE
final_smape = get_final_smape(smape_rougher, smape_final) 
print('Итговая оценка SMAPE для модели линейной регрессии', round(final_smape, 2)*-1)
```

*На тренировочной выборке Итоговое значенине SMAPE модели линейной регресии составило 10.1*

### Обучение и выбор модели Случайного леса

```python
%%time
model_rouger_rf = RandomForestRegressor(random_state=12345)

params = { 'max_depth': [x for x in range(1, 20, 5)],
               'n_estimators': [x for x in range(10, 250, 25)]
             }

grid_rouger_rf = GridSearchCV(model_rouger_rf, params, cv=3, scoring=smape_scorer, n_jobs=-1, verbose=10)
grid_rouger_rf.fit(features_train, target_train_rouger)

print('Параметры лучшей модели: ',grid_rouger_rf.best_params_)
print('SMAPE для модели на стадии флотации', (round(grid_rouger_rf.best_score_, 2))*-1)
```

```python
%%time
model_final_rf = RandomForestRegressor(random_state=12345)

params = { 'max_depth': [x for x in range(1, 20, 5)],
               'n_estimators': [x for x in range(10, 250, 25)]
             }

grid_final_rf = GridSearchCV(model_final_rf, params, cv=3, scoring=smape_scorer, n_jobs=-1, verbose=10)
grid_final_rf.fit(features_train, target_train_final)

print('Параметры лучшей модели: ', grid_final_rf.best_params_)
print('SMAPE для модели на финальной стадии', (round(grid_final_rf.best_score_, 2))*-1)
```

```python
final_smape_rf = get_final_smape(grid_rouger_rf.best_score_, grid_final_rf.best_score_)
print('Итговая оценка SMAPE для модели случайного леса', round(final_smape_rf, 2)*-1)
```

*Итоговая оценка SMAPE модели случайного леса составила 9.34, наилучшие показатели  предсказания на стадии флотации дала модель с количеством деревьев 35 и макимальной глубиной ветвоения 6, для стадии финальной очистки - количество деревьев 85 и максимальной глубиной ветвления 1.*
