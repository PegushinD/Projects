### Обработка пропусков данных

#### В столбцах name, year_of_release, genre, пропущеных значений всего несколько строк, на результать анализа такое количество не повлияет, удалим строки с пропусками в этих столбцах

```python
pd.set_option('display.max_rows', None)
df.loc[(df['name'].isna()) | (df['year_of_release'].isna()) | (df['genre'].isna())]
```

```python
df.info()
```

Это может возникнуть из-за слишком раннего выхода игры когда рейтинги еще не собирались или наоборот игра может быть слишком новой, так же заметил связь  отсутствия рейтинга у игр с очень низкими продажами, то есть не получившими большого распространения.* 

*Пропуски в рейтинге и оценках оставим как есть, заполнять их средними значениями нельзя, в дальнейшем будем использовать данные "как есть".*

### Посмотрите, как менялись продажи по платформам. Выберите платформы с наибольшими суммарными продажами и постройте распределение по годам. За какой характерный срок появляются новые и исчезают старые платформы?


#### Платформы с наибольшими суммарными продажами

```python
#выберем платформы с наибольшими суммарными продажами и сохраним список в новый датафрейм он нам в дольнейшем пригодится
platforms_sales = (
    df.pivot_table(index = 'platform', values = 'world_sales', aggfunc = 'sum')
    .sort_values(by = 'world_sales', ascending = False)
)
platforms_sales = platforms_sales.reset_index()
```

```python
top_platform_sales = platforms_sales.loc[:7]
top_platform_sales
```

#### Распределение продаж по годам для самых популярных платформ

```python
#Построим график распределния продаж по годам для 7 самых популярных платформ
(
    df.query('year_of_release >= 2002 and platform in @top_platform_sales["platform"]')
    .pivot_table(index='year_of_release', columns = 'platform', values='world_sales', aggfunc='sum')
    .plot(figsize=(20,8))
)
```

#### Жизненный цикл платформ

```python
life_cycle = (
    df.query('year_of_release >= 1995 and platform in @top_platform_sales["platform"]')
    .pivot_table(index='platform', values='year_of_release', aggfunc=('min', 'max'))
)
                
```

```python
life_cycle['year'] = life_cycle['max'] - life_cycle['min']
display(life_cycle.sort_values(by='min'))
display('Средняя продолжительность жизни платформы:', life_cycle['year'].mean())
display('Медианная продолжительность жизни платформы:', life_cycle['year'].median())
```

*Для дальнейшего анализа мы оставили 7 самых продаваемых в мире платформ. По общей сумме продаж безусловным лидером является PS2 но пик популярности этой платформы пришелся на далекий 2004 год, но даже самые популярны платформы более позднего времени не поднимались до уровня PS2. Средняя продолжительность "жизни" игровой платформы составляет 9 лет, новые платформы появляются с периодичностью 4-5 лет*

### Влияние на продажи внутри  платформы PS3 отзывов пользователей и критиков. 

```python
#построим диаграммы рассеяния по зависмости продаж по играм платформы от оценки критиков и пользователей
df_ps3 = df.loc[df['platform'] == 'PS3']
df_ps3.plot(kind = 'scatter', x='critic_score', y='world_sales')

df_ps3.plot(kind = 'scatter', x='user_score', y='world_sales')
```

```python
#посчитаем коэффициент корреляции Пирсона оценок критиков и мировых продаж
df_ps3['world_sales'].corr(df_ps3['critic_score'])
```

```python
#посчитаем коэффициент корреляции Пирсона оценок пользователей и мировых продаж
df_ps3['world_sales'].corr(df_ps3['user_score'])
```

*Связь продаж PS3 с оценками критиков выражена сильнее чем с оценок пользователей, продажи по данной платформе в большей степени коррелируют с оценками критиков.*

## Проверка гипотез

### Проверим гипотезу: средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Н_0: Среднйи пользовательский рейтинг платформы Xbox One = Среднйи пользовательский рейтинг платформы PC

H_1: Среднйи пользовательский рейтинг платформы Xbox One ≠ Среднйи пользовательский рейтинг платформы PC

Так как проверяем гипотезу о равенстве средних двух независимых друг от друга совокупностей, применять будем метод scipy.stats.ttest_ind(), параметр  equal_var зададим как False, порого статистической значимости установим как общепринятые 5%

Для получения корректных значений пропуски в выборках удалим

```python
rating_xone = df.loc[(df['platform'] == 'XOne')&(df['year_of_release']< 2016) , 'user_score'].dropna()
rating_pc = df.loc[(df['platform'] == 'PC')&(df['year_of_release']< 2016)&(df['year_of_release']>= 2013), 'user_score'].dropna()
alpha = .05

results = st.ttest_ind(rating_xone, rating_pc, equal_var = False)

display(results.pvalue)

if results.pvalue < alpha:
    display('Отвергаем нулевую гипотезу')
else:
    display('Не получилось отвергнуть нулевую гипотезу')
```

### Проверим гипотезу: средние пользовательские рейтинги жанров Action  и Sports разные

Н_0: Среднйи пользовательский рейтинг жанра Action  = Среднйи пользовательский рейтинг жанра Sports

H_1: Среднйи пользовательский рейтинг жанра Action ≠ Среднйи пользовательский рейтинг жанра Sports

```python
rating_action = df.loc[(df['genre'] == 'Action'), 'user_score'].dropna()
rating_sports = df.loc[(df['genre'] == 'Sports'), 'user_score'].dropna()
alpha = .05

results = st.ttest_ind(rating_action, rating_sports, equal_var = False)

display(results.pvalue)

if results.pvalue < alpha:
    display('Отвергаем нулевую гипотезу')
else:
    display('Не получилось отвергнуть нулевую гипотезу')
```

```python
# check
# круги + категория другие
def graph (df, year, region, name, axes):
    
    df = df.query('year_of_release >= @year')
    
    sales = df.pivot_table(index='platform', 
                           values=region, 
                           aggfunc='sum').nlargest(5, region)
    
    sales = sales.reset_index()
    
    sales = (
            sales.append({'platform': 'Other', region: df[region].sum() 
                       - sales[region].sum()}, ignore_index= True)
         )
    
    
    sales.columns = ['platform', 'sales']
      
    labels_c=sales.platform
    colours = {'Wii':'C0', 'NES':'C1', 'GB':'C2', 'DS':'C3', 'X360':'C4', 
    'PS3':'C5', 'PS2':'C6', 'SNES':'C7', 'GBA':'C8',
               'PS4':'steelblue', '3DS':'orange', 
               'N64':'C11', 'PS':'C12', 'XB':'C13', 'PC':'C14', '2600':'C15', 'PSP':'C16', 
               'XOne':'C17',
               'WiiU':'C18', 'GC':'C19', 'GEN':'C20', 'DC':'C21', 'PSV':'C22', 
               'SAT':'C23', 'SCD':'C24', 'WS':'C25', 'NG':'C26', 
               'TG16':'C27', '3DO':'C28', 'GG':'C29', 'PCFX':'C30', 'Other':'darkred'}
   
    sales.plot(kind='pie',
               y="sales",
               
               autopct='%1.0f%%',
               wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
               textprops={'size': 'x-large'}, 
               labels= labels_c,
               colors=[colours[key] for key in labels_c],
               legend=False, 
               title = f"Популярность платформ в {name} ", 
               ax = axes).set(ylabel='')
    
    plt.tight_layout()
```
